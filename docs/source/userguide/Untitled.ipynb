```python
%matplotlib inline

from pylab import *

phi = lambda n,x: sin( n*pi*(x+1/2.))

x =arange(-0.51,0.51,.01)

phi1 = phi(1,x)
phi2 = phi(2,x)
phi3 = phi(3,x)
phi4 = phi(4,x)

Q = 2.*phi1**2  + 2.*phi2**2  + 2.*phi3**2  
Q2 = 2.*phi1**2  + 2.*phi2**2  + phi3**2 + phi4**2
_ = plot(x, Q)
_ = plot(x, Q2)
```
